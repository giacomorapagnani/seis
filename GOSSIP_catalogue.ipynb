```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
from obspy import UTCDateTime
from obspy.core.event import Catalog
from obspy.core.event import Event
from obspy.core.event import Origin
from obspy.core.event import Magnitude
```

# Read and create dictionary

```python
directory = os.fsencode('flegrei_2014_2023_GOSSIP/events/')
cat_all=pd.DataFrame()

print('reading and merging:')
for file in os.listdir(directory):
    filename = os.fsdecode(file)

    if filename.endswith(".csv"): 
        print(filename)
        cat=pd.read_csv('flegrei_2014_2023_GOSSIP/events/'+filename)
        cat=cat.drop(labels=['Area','Type'],axis=1)
        cat['Time'] = cat['Time'].apply(UTCDateTime)
        cat_all=pd.concat( [cat_all,cat] ,ignore_index=True)
    else:
        continue

for location,id_event in enumerate(cat_all['#EventID']):
    if cat_all['Latitude'].isna().loc[location] and cat_all['MagType'].isna().loc[location]:
        cat_all=cat_all.drop(index=location)
print('into: \nevents_all.csv')


cat_all.to_csv('/flegrei_2014_2023_GOSSIP/events_all.csv')
```

# create catalogue obspy

```python
cat_obs=Catalog()
for ind,id_number in enumerate(cat_all['#EventID']):
    
    event=Event()
    event.resource_id=str(id_number)
    event.comments=cat_all['Level'].iloc[ind]
    
    origin=Origin()
    origin.time=cat_all['Time'].iloc[ind]

    magnitude=Magnitude()

    try:
        origin.latitude=cat_all['Latitude'].iloc[ind]
    except:
        None
    
    try:
        origin.longitude=cat_all['Longitude'].iloc[ind]
    except:
        None
    
    try:
        origin.depth=cat_all['Depth/km'].iloc[ind]
    except:
        None
    
    try:
        magnitude.mag=cat_all['Magnitude'].iloc[ind]
    except:
        None

    try:
        magnitude.mag_errors=cat_all['MagErr'].iloc[ind]
    except:
        None

    try:
        magnitude.magnitude_type=cat_all['MagType'].iloc[ind]
    except:
        None

    event.origins.append(origin)
    event.magnitudes.append(magnitude)
    cat_obs.events.append(event)
```
